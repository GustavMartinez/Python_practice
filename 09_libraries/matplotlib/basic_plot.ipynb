```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# plot

year = [1950,1970,1990,2010]
pop = [2.5,3.6,5.2,6.9]

plt.plot(year, pop)
plt.grid(alpha=0.2)
plt.show()
```

```python
# Scatter plot

year = [1950,1970,1990,2010]
pop = [2.5,3.6,5.2,6.9]

plt.scatter(year, pop) # x = year; y = pop
plt.grid(alpha=0.3)
plt.show()
```

```python
# Import data gapminder.csv

DATA_PATH = '/home/gustavo/Documents/Study/005_Python_General/github/python_practice/09_libraries/00_data/gapminder.csv'

gdp_data_2007 = pd.read_csv(DATA_PATH)

gdp_data_2007.head(3)
```

## Exploring correlations

```python
plt.scatter(gdp_data_2007['gdp_cap'], gdp_data_2007['life_exp']) # Passing the data
plt.xscale('log') # Show data in logarithmic scale
plt.xlabel("GDP cap")
plt.ylabel("Life expectancy")
plt.grid(alpha=0.2)
plt.show()
```

```python
plt.scatter(gdp_data_2007['population'], gdp_data_2007['life_exp'])
plt.ylabel("Life Expectancy")
plt.xlabel("Population")
plt.show()
```
