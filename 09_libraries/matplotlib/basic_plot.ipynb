```python
import matplotlib.pyplot as plt
```

```python
# plot

year = [1950,1970,1990,2010]
pop = [2.5,3.6,5.2,6.9]

plt.plot(year, pop)
plt.grid(alpha=0.2)
plt.show()
```

```python
# Scatter plot

year = [1950,1970,1990,2010]
pop = [2.5,3.6,5.2,6.9]

plt.scatter(year, pop) # x = year; y = pop
plt.grid(alpha=0.3)
plt.show()
```
