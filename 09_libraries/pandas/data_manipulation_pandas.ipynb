```python
import pandas as pd
```

```python
PATH_DATA = '/home/gustavo/Documents/Study/005_Python_General/github/python_practice/09_libraries/00_data/gapminder.csv'
df = pd.read_csv(PATH_DATA, delimiter=',', index_col=0)

df.head(3)
```

### Sorting by one variable

```python
# Sorting

df.sort_values('cont').head(8) # Apply the method to a specific column
```

```python
# Sorting values by the population column, with ascending=False

df.sort_values('population', ascending=False).head(6)
```

```python
df.sort_values('life_exp').head(6) # min to max or descending order
```

### Sorting by multiple variables

```python
df.sort_values(['cont', 'life_exp']) # The dataset is sorted first by continent, and then by life_expectancy.
```

```python
df.sort_values(['cont', 'life_exp'], ascending=[True, False]) # This is also posible
```

### Subseting by conditions

```python
df[df["life_exp"] > 80]
```

```python
life_more_than_80 = df['life_exp'] > 80
in_europe = df['cont'] == 'Europe'

df[life_more_than_80 & in_europe]
```

```python
# The last result can be done like:

df[(df['life_exp'] > 80) & (df['cont'] == 'Europe')].sort_values('life_exp', ascending=False)
```

### Subseting using .isin()

```python
italy_spain = df['country'].isin(['Italy', 'Spain'])
df[italy_spain]
```
