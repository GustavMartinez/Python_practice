## Duplicates

```python
# Drop duplicate store/type combinations

store_types = sales_df.drop_duplicates(subset=(['store','type']))
store_types
```

```python
# Drop duplicates store/department combinations

store_depts = sales_df.drop_duplicates(subset=(['store','department']))
store_depts.head()
```

```python
# Subset the rows where is_holiday is True and drop duplicate dates

holiday_dates = sales_df[sales_df['is_holiday']].drop_duplicates('date')
holiday_dates
```

## Counting categorical variables

```python
store_counts = store_types['type'].value_counts()
store_counts
```

```python
store_props = store_types['type'].value_counts(normalize=True)
store_props
```

```python
# Count the number of stores of each type

store_counts = store_types['type'].value_counts()
store_counts
```

```python
# Get the proportion of stores of each type

store_props = store_types['type'].value_counts(normalize=True)
store_props
```

```python
# Count the number of stores for each department and sort

dept_counts_sorted = store_depts['department'].value_counts().sort_values(ascending=False)
dept_counts_sorted
```

```python
# Get the proportion of stores in each department and sort

dept_props_sorted = store_depts['department'].value_counts(normalize=True).sort_values(ascending=False)
dept_props_sorted
```
